## main difference between eclat model is we visualize results with only on the basis of support

```python
from apyori import apriori
rules =apriori(transactions=transactions, min_support=0.0005, min_confidence = 0.05, min_lift=2, min_length=2, max_length=2)
results = list(rules)
results

```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    
    return list(zip(lhs, rhs, supports))
ordered_results = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support'])
```

```python
ordered_results.head(10)
```

```python
ordered_results.nlargest(n=10,columns='Support')
```
